```python
from sklearn.tree import DecisionTreeClassifier
```

```python
df=pd.read_csv(r"C:\Users\Admin\Downloads\Renewable_Energy_Adoption.csv")
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
#feature and target selection
X=df[['carbon_emissions', 'energy_output', 'renewability_index',
       'cost_efficiency']]
y=df['adoption']
```

```python
#train test split
Xtrain,Xtest,ytrain,ytest=train_test_split(X,y,train_size=0.8,random_state=42)
```

```python
#instantiate the model
model=DecisionTreeClassifier(max_depth=3,random_state=42)
```

```python
#train the model
model.fit(Xtrain,ytrain)
```

```python
#prediction
y_pred=model.predict(Xtest)
y_pred
```

```python
#evaluation
acc=accuracy_score(ytest,y_pred)
print(f'the accuracy score for DTC is {acc*100:.2f}%')
```

```python
#confution matrix
cm=confusion_matrix(ytest,y_pred)
sns.heatmap(cm,annot=True,cmap='Purples',
           xticklabels={'Non Adoption','Adoption'},
           yticklabels={'Non Adoption','Adoption'})
plt.title('heatmap of confusion matrix')
plt.xlabel('predicted values')
plt.ylabel('actal values')
plt.show()
```

```python
from sklearn.tree import plot_tree
```

```python
plt.figure(figsize=(12,10))
plot_tree(model,filled=True,feature_names=X.columns,
         class_names=['Non Adoption','Adoption'])
plt.show()
```

```python
#classification report
cr=classification_report(ytest,y_pred)
print(cr)
```
